```python
from matplotlib import Image
import matplotlib.pyplot as plt

image = Image.open("images/steak.jpeg")
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

image = Image.open("images/steak.jpeg")
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
from PIL import Image

image = Image.open("images/steak.jpeg")
image.thumbnail((224, 224))
image.save("images/steak_resized.jpeg")
```

```python
import matplotlib.pyplot as plt

plt.imshow(image)
plt.axis("off")
plt.show()
```
